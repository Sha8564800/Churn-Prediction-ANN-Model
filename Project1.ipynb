```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder_gender=LabelEncoder()
df['Gender']=label_encoder_gender.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
df['Geography'].unique()
```

```python
df.head()
```

```python
#One hot encoding of geography column in dataset
#We can also use label encoding but it will encode france as 1 and germany as 2 and spain as 3 but as ANN is all about numerical calculation so germany can be treated greater value than france

from sklearn.preprocessing import OneHotEncoder
ohe_geography=OneHotEncoder()
geo_encoder=ohe_geography.fit_transform(df[['Geography']])
geo_encoder
```

```python
geo_encoder.toarray()
```

```python
ohe_geography.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df=pd.DataFrame(geo_encoder.toarray(),columns=ohe_geography.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
#Combining one hot encoder columns with the original data
df=pd.concat([df.drop('Geography',axis=1),geo_encoded_df],axis=1)
```

```python
df.head()
```

```python
import pickle
```

```python
#Save the encoders and scalars
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('ohe_geography.pkl','wb') as file:
    pickle.dump(ohe_geography,file)    
```

```python
df.head()
```

```python
#Devide our dataset into dependent and independent features
#Independent feature
x=df.drop('Exited',axis=1)
#dependent feature
y=df['Exited']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#Saving the scaler in pickle file
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

## Now our data is ready and on this df dataset we will train our ANN

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential      #For sequential network
from tensorflow.keras.layers import Dense,Input           #for making the number of neuron layers
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
x_train.shape[1]
```

```python
# Build our ANN Model

model=Sequential([
    Input(shape=(x_train.shape[1],)),
    Dense(64, activation='relu'),    #First hidden layer connected with input layer
    Dense(32,activation='relu'),     #Second hidden layer
    Dense(1,activation='sigmoid')   #output layer
])
```

```python
model.summary()
```

```python
#Compile the model for the forward and backward propagation

import tensorflow

opt=tensorflow.keras.optimizers.Adam(learning_rate=0.001)
loss=tensorflow.keras.losses.BinaryCrossentropy()

model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

```python
#Set up the tensorboard

from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
log_dir = "logs/fits/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
#Set up Early Stopping so that if the model is trained in less epochas than specified then it should stop early

early_stopping_callback=EarlyStopping(monitor='val_loss',patience=15,restore_best_weights=True)
```

```python
#Train the model

history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,
                  callbacks=[tensorflow_callback,early_stopping_callback]
                )
```

```python
#Save the model file
model.save('model.h5')
```
